```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load in the arrests_by_day data
arrests_by_day = pd.read_csv("arrests_by_day.csv")
```

```python
# fixing the index column
arrests_by_day = arrests_by_day.set_index("Unnamed: 0")
```

```python
arrests_by_day_test = arrests_by_day.copy()

# grouping by week of data
dates = pd.to_datetime(arrests_by_day_test["Date"])
arrests_by_week_test = arrests_by_day_test.assign(Date = dates).groupby(pd.Grouper(key="Date", freq = "1W")).sum()

# reassigning proportion column to reflect proprtions of non citizenship country deportations to citizenship country deportations
arrests_by_week_test["Home country proportion"] = arrests_by_week_test["Deportations elsewhere"]/arrests_by_week_test["Deportations to home country"]

# changes the index from the date to a new default index column
arrests_by_week_test = arrests_by_week_test.reset_index(drop=False)

# drops the last incomplete week of data so the graphs only use full weeks of data
arrests_by_week_test = arrests_by_week_test.drop(arrests_by_week_test.tail(1).index)
```

```python
# plots count of arrests per week in a line plot
x = arrests_by_week_test["Date"]
y = arrests_by_week_test["Arrests"]

plt.plot(x, y)
plt.xlabel('Date')
plt.xticks(ticks = [19631., 19723., 19814., 19905., 19997., 20089., 20179., 20270.], 
           labels = ["Oct 2023", "Jan 2024", "Apr 2024", "Jul 2024", "Oct 2024", "Jan 2025", "Apr 2025", "Jul 2025"],
           rotation = 20)
plt.ylabel('Arrests')
plt.title('Arrests per week (Sep 2023 to Aug 2025)')
plt.axvline(x = 20239.66, color = 'b', label = 'six months into Trump presidency', ls = "--")
plt.axvline(x = 20089, color = 'm', label = 'Trump presidency', ls = ":")
plt.legend()
plt.grid(True)
plt.show()
```
